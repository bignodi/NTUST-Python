```python
#### str advance !!

###############################
## .format 字符串模版
###############################

name = "Jeff"
w1 = "Hi, I'm {} "

print(w1)
print(w1.format(name))

str1 = "one"
str2 = "two"
wordList = "Hi, 1st str :{}, 2nd str: {}"
print(wordList.format(str1,str2))
print(wordList.format(str2,str1))


print("="*20)

###############################
## [:] sub string 子字串
###############################

w2 = "This is a book"
print("取第10個之後 [10:]: ", w2[10:])
print("取3到12[3:12]: ", w2[3:12])
print("取3到倒數第二個字元[3:-2]: ", w2[3:-2])

print("="*20)
###############################
## [.] split, 分割字串, 回傳陣列
###############################

w3 = "This is a book."
w4 = "apple,banana,Cat"
print(w3.split(" "))
print(w4.split(" "))
print(w4.split(","))



print("="*20)
###############################
## .join, 合併字串, 回傳陣列
###############################

w5 = "".join(["One","Two","Three"])
w51 = ",".join(["One","Two","Three"])
w52 = "___".join(["One","Two","Three"])
w53 = "_______".join(["One","Two","Three","Four","Five"])

print(w5)
print(w51)
print(w52)
print(w53)




print("="*20)
###############################
## .strip, 
###############################

w6 = "                     abcde"
print(w6)
print(w6.strip())



print("="*20)
###############################
## .startswith 檢查是否是xxx 開頭, 回傳true/false
###############################
w7 = "Avenger Assemble!"
print("檢查字串Avenger 存在嗎? ", w7.startswith("Avenger"))
print("檢查字串test 存在嗎? ", w7.startswith("test"))
print(type(w7.startswith("Avenger")))




print("="*20)
###############################
## .replace 取代
###############################

w8 = "Green tea."
print(w8.replace("Green","Red"))
print(w8.replace("Blue","Red"))


print("="*20)
###############################
## .find  尋找字串位置 (-1 表不存在)
###############################

w9 = "Today is a rainy day !"
print(w9)
print("is 在", w9.find("is"))
print("are 不存在", w9.find("are"))
```

```python
##### 編碼
test1 = "Jeff"
print(test1)
print(test1.encode("utf8"))

print("="*20)

test2 = "周杰倫"
print(test2)
print(test2.encode("utf8"))

print(type(test2))
print(type(test2.encode("utf8")))


print("="*20)
test3 = b'\xe5\x91\xa8\xe6\x9d\xb0\xe5\x80\xab'

print(test3)
print(type(test3))
print(test3.decode("utf8"))

## 一定要加 b'\x..\x..' 才會被解讀成byte, 否則會變成一般string 變成亂碼
test4 = '\xe5\x91\xa8\xe6\x9d\xb0\xe5\x80\xab'
print(test4)
print(type(test4))

## \x 表示ASCII 開頭
test5 = '\x61'
print(test5)
print(type(test5))

```
